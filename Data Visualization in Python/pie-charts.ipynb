```python
#Importing packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing from an online source
df = pd.read_csv('https://www.fhwa.dot.gov/bridge/nbi/2022/delimited/NY22.txt', sep=',', engine='python')
df
```

```python
#Grouping the data by Bridge Condition and counting off the birdges by 'P', 'F', and 'G'
df.groupby("BRIDGE_CONDITION")["STATE_CODE_001"].count()
```

You already know how to do the above - if you don't, look back at the morning wake-up activity from 7/21 (https://colab.research.google.com/drive/1UdT3Bhx-4KLAjEF_d0ue0V4l0gbh3VbB?usp=sharing). 

One way of doing this problem is what we went over on Thursday - since we don't recognize what this new object is (the groupby and count object), we must convert into a format that the pie chart functions can understand.

The other option is (by which we bypass a lot of the code) if we can find a way to pass the entire object as it is through the pie chart without calling it. 

```python
#Creating a pie-chart using the pandas dataframe
df.groupby("BRIDGE_CONDITION")["STATE_CODE_001"].count().plot.pie(figsize=(10,10), autopct='%1.1f%%')

#The figsize only helps largen the image. You can ignore it if you want. The autopct attributes helps you display percentages. 
```
