```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

boat = pd.read_csv('NYC_Ferry_Ridership.csv', sep=',', engine='python')
boat
stop_groups = boat.groupby('Stop')['Boardings'].count()
stop_groups
#When I print 'stop_groups', it gives me the result as expected but this object is a series and not a DataFrame. You have to convert into a DataFrame by reindexing it.
#Incorrect method:
#df.insert(0, boat['Stop'], True)
#stop_groups was still a series at this point and hence was not hashable. You must index it before you do anything else. 

#Correct Solution
df = boat.groupby('Stop')['Boardings'].count().reset_index()
df
#Printing df gives me a nice looking dataframe to work with (that has been indexed). Notice the difference when you try and just print df from the incorrect solution and when you print it from my code.
#After you do this, you can go ahead and plot as you were doing earlier.
plt.plot(df['Stop'],df['Boardings'])
plt.title('title name')
plt.xlabel('xAxis name')
plt.ylabel('yAxis name')
plt.show()

#Here is the link to the stackoverflow page that helped me troubleshoot if you are interested: 'https://stackoverflow.com/questions/10373660/converting-a-pandas-groupby-output-from-series-to-dataframe'
```
